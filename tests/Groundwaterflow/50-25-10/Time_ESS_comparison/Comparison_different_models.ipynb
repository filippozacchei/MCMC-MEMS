# Compare the different Time / ESS of the different sampling strategies 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the data for noise $\sigma^2$ = 0.001

```python
Time_ESS_FOM_001 = np.load('./recorded_values/FOM_time_ess_001.npy')
Time_ESS_MDA_MF2step_001 = np.load('./recorded_values/MDA_MF_2step_time_ess_001.npy')
Time_ESS_MDA_MF1step_001 = np.load('./recorded_values/MDA_MF_1step_time_ess_001.npy')
Time_ESS_MF2step_001 = np.load('./recorded_values/MF_2step_time_ess_001.npy')
Time_ESS_MF1step_001 = np.load('./recorded_values/MF_1step_time_ess_001.npy')
```

```python
# Combine data into a single list
data = [Time_ESS_FOM_001, Time_ESS_MDA_MF2step_001, Time_ESS_MDA_MF1step_001, 
        Time_ESS_MF2step_001, Time_ESS_MF1step_001]

# Labels for each dataset
labels = ['FOM', 'MDA with MF 2-step', 'MDA with MF 1-step', 'Only MF 2-step', 'Only MF 1-step']

# Set the color palette to "crest"
sns.set_palette("BuGn")

# Create the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, palette="crest", alpha=0.7)

# Set title and labels
plt.title('Time/ESS distribution for Different Approaches', fontsize=14)
plt.ylabel('Time/ESS', fontsize=12)
plt.xticks(ticks=range(5), labels=labels, fontsize=10)

# Show the plot
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
